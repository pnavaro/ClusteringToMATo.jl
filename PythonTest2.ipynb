```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import time
from sklearn import cluster, datasets
%matplotlib inline
```

```python
from scipy import sparse
```

```python
TD = np.loadtxt(".\\FCPS\\01FCPSdata\\TwoDiamonds.txt");
```

```python
n=2;
```

```python
sp1=cluster.SpectralClustering(n_clusters=n,eigen_solver='arpack')
```

```python
tstart=time.time()
data1 = cluster.SpectralClustering.fit_predict(sp1,TD);
tend=time.time()
print (tend-tstart)
```

```python
A=[[] for i in range(n)]
for i in range(len(data1)):
    for j in range(n):
        if data1[i]==j:
            A[j].append(i)
```

```python
plt.figure(figsize=(6, 4),dpi=100)
plt.axis('equal')
for i in range(n):
    plt.scatter(TD[A[i],0],TD[A[i],1],linewidths=0.1,edgecolors='black')
plt.savefig("TwoDiamonds_Sp.png")
```
